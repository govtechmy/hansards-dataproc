```python
def not_invisible_rect(obj):
    if obj["object_type"] != "rect":
        return True
    if isinstance(obj["non_stroking_color"], int) or isinstance(obj["non_stroking_color"], float):
        return obj["non_stroking_color"] < 0.9
    return not (min(obj["non_stroking_color"]) > 0.9)
```

```python
# Edit the DR/DN, year, date and page number
pdf = pdfplumber.open("data/debug/KKDR-15112023.pdf")
p = pdf.pages[10-1]
p.filter(not_invisible_rect).extract_tables({"snap_x_tolerance":9})
```

```python
p.filter(not_invisible_rect).to_image().debug_tablefinder({"snap_tolerance":9})
```
